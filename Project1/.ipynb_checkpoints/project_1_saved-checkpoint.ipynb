In 2018 we have a problem, because we have about three days without data. It is weekend data. This we cannot interpolate but we can assume that from one weekend in February to the next the conditions don't change too much. We can take the mean of the data of the week before and after, which implicitly assumes that there are no big fluctiations on a weekend and the power consumption changes gradually. Filling it with data from the previous year is a stretch because the power consumption will vary depending on the outside tempereture. This temperature changes gradually over a week but has no connection to the temperature on year earlier.

We can go on and add more features such as holidays or weekdays. Probably the power consumption will be lower on a sunday than on a thursday. On Christmas it will be lower than on regular workdays

```python
def add_holiday_to_dataframe(dataframe, inplace = False):
  holidays_pt = holidays.PT(subdiv='11', years=[2017,2018,2019])
  if inplace == True: new_frame = dataframe
  else: new_frame = dataframe.copy()
  #print(new_frame.index)
  new_frame['Holiday'] = new_frame.index
  new_frame['Holiday'] = new_frame['Holiday'].dt.date.isin(holidays_pt)
  return new_frame
def add_weekday_to_dataframe(dataframe, inplace = False):
  if inplace == True: new_frame = dataframe
  else: new_frame = dataframe.copy()

  new_frame['Weekday'] = new_frame.index
  new_frame['Weekday'] = new_frame.index.dayofweek
  return new_frame
def add_dayhour_to_dataframe(dataframe, inplace = False):
  if inplace == True: new_frame = dataframe
  else: new_frame = dataframe.copy()

  new_frame['Hour'] = new_frame.index
  new_frame['Hour'] = new_frame.index.hour
  return new_frame
def add_month_to_dataframe(dataframe, inplace = False):
  if inplace == True: new_frame = dataframe
  else: new_frame = dataframe.copy()

  new_frame['Month'] = new_frame.index
  new_frame['Month'] = new_frame.index.month
  return new_frame
def power_minus_n(dataframe, n=1, inplace = False):
  if inplace == True: new_frame = dataframe
  else: new_frame = dataframe.copy()

  new_frame['Power -'+str(n)] = new_frame['Power_kW'].shift(1)
  return new_frame
# Code for testing the holiday and weekday functions
#copy = power_data.copy()
#copy = (
#  copy.pipe(add_holiday_to_dataframe).
#  pipe(add_weekday_to_dataframe)
#)
#print(copy)
```

Adding more features than Power -1 (Power -2,-3,...,-24,...) does not improve the results

```python
augumented_data = all_data.copy()
augumented_data = (
  augumented_data.pipe(add_holiday_to_dataframe)
  .pipe(add_weekday_to_dataframe)
  .pipe(add_dayhour_to_dataframe)
  .pipe(add_month_to_dataframe)
  .pipe(power_minus_n,n=1)
).dropna()
augumented_data.head()
#augumented_data.info()
```

## Regression

Before we use regression models we want to analyse the seasonal components of the time series data

```python
import statsmodels.api as sm
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from pandas.plotting import lag_plot
from sklearn import  metrics
```

```python
def ml_metrics(pred, check):
    MAE=metrics.mean_absolute_error(check,pred) 
    MBE=np.mean(check-pred) 
    MSE=metrics.mean_squared_error(check,pred)  
    RMSE= np.sqrt(metrics.mean_squared_error(check,pred))
    cvRMSE=RMSE/np.mean(check)
    NMBE=MBE/np.mean(check)
    r2 = metrics.r2_score(check,pred)
    print('R2     :',r2)

    print('MAE    :', MAE)
    print('MBE    :', MBE)
    print('MSE    :', MSE)
    print('RMSE   :', RMSE)
    print('cvRMSE :', cvRMSE)
    print('NMBE   :', NMBE)
```

```python
def plotseasonal(res, axes ):
    #code from https://stackoverflow.com/questions/45184055/how-to-plot-multiple-seasonal-decompose-plots-in-one-figure
    res.observed.plot(ax=axes[0], legend=False)
    axes[0].set_ylabel('Observed')
    res.trend.plot(ax=axes[1], legend=False)
    axes[1].set_ylabel('Trend')
    res.seasonal.plot(ax=axes[2], legend=False)
    axes[2].set_ylabel('Seasonal')
    res.resid.plot(ax=axes[3], legend=False)
    axes[3].set_ylabel('Residual')

fig, axes = plt.subplots(ncols=2, nrows=4, sharex=True, figsize=(12,5))

freq = 24
res = sm.tsa.seasonal_decompose(augumented_data['Power_kW'],
                                period=freq,
                                model='multiplicative')
plotseasonal(res, axes[:,0])


freq = 24*7
res = sm.tsa.seasonal_decompose(augumented_data['Power_kW'],
                                period=freq,
                                model='multiplicative')
plotseasonal(res, axes[:,1])


plt.show()
```

```python

fig, ax = plt.subplots(2,2,figsize=(10,10))
fig.tight_layout(pad=5.0)

autocorrelation_plot(augumented_data['Power_kW'], ax=ax[0,0])

plot_acf(augumented_data['Power_kW'], ax=ax[0,1])

plot_pacf(augumented_data['Power_kW'],lags=24, ax=ax[1,0])

lag_plot(augumented_data['Power_kW'], ax=ax[1,1])
plt.show()
```

### Test train split

We first split the dataset into train, and validation data. The test data will be 2019 data.

```python
print(augumented_data.columns)
```

```python
def split_features(featurelist,split=0.75):
    df = augumented_data[featurelist]
    n = len(df)
    first = int(n*split)

    df_train = df.iloc[:first]
    df_val = df.iloc[first:]

    t_train = df.index[:first]
    t_val = df.index[first:]

    Z_train = df_train.values
    Z_val = df_val.values

    Y_train = df_train['Power_kW'].values
    Y_val = df_val['Power_kW'].values

    X_train = df_train.drop(columns=['Power_kW']).values
    X_val = df_val.drop(columns=['Power_kW']).values

    return X_train, X_val, Y_train, Y_val, t_train, t_val
```

### Autoregressive

In the Autoregressive I chose a window of 2 because in the partial autocorrelation plot we have two spikes in the beginning.

```python
from statsmodels.tsa.ar_model import AutoReg

# train autoregression
window = 2
model = AutoReg(Y_train, lags=window)
#model=ARIMA(train, order=(1, 1,1))
model_fit = model.fit()

print(model_fit.summary())
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

coef = model_fit.params
# walk forward over time steps in test
history = Y_train[len(t_train)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()
for t in range(len(Y_val)):
    length = len(history)
    lag = [history[i] for i in range(length-window,length)]
    yhat = coef[0]
    for d in range(window):
        yhat += coef[d+1] * lag[window-d-1]
    obs = Y_val[t]
    predictions.append(yhat)
    history.append(obs)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(t_val,Y_val,label='Data')
plt.plot(t_val,predictions, color='red',label='Predictions')
plt.legend()

plt.subplot(1,2,2)
plt.scatter(Y_val,predictions)
plt.show()
ml_metrics(predictions,Y_val)
```

The windowsize of 2 gives the best results. This is not suprising looking at the partial autocorrelation plot.

### Random Forest

In random forests it is not as important to perform a feture selection, so I will use all the features that I have accumulated so far.

```python
augumented_data.columns
```

```python
from sklearn.ensemble import RandomForestRegressor

#featurelist_LR = ['Power_kW','temp_C','Power -1','Hour','Weekday','Holiday', 'windGust_m/s']
featurelist_RF = ['Power_kW', 'HR', 'pres_mbar', 'rain_mm/h', 'solarRad_W/m2', 'temp_C',
       'windGust_m/s', 'windSpeed_m/s', 'Holiday', 'Weekday', 'Hour', 'Month',
       'Power -1']

X_train, X_val, Y_train, Y_val, t_train, t_val = split_features(featurelist_RF)

parameters = {'bootstrap': True,
              'min_samples_leaf': 5,
              'n_estimators': 200, 
              'min_samples_split': 5,
              'max_features': 100,#'sqrt',
              'max_depth': 50,
              'max_leaf_nodes': None}
RF_model = RandomForestRegressor(**parameters)
#RF_model = RandomForestRegressor()
RF_model.fit(X_train, Y_train)

RF_val_pred = RF_model.predict(X_val)
RF_train_pred = RF_model.predict(X_train)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)

plt.plot(t_val,Y_val)
plt.plot(t_val,RF_val_pred)

plt.subplot(1,2,2)
plt.scatter(Y_val,RF_val_pred)

plt.show()

print("===  Training  ===")
ml_metrics(RF_train_pred,Y_train)
print("=== Validation ===")
ml_metrics(RF_val_pred,Y_val)
```

### Neural Network

```python
import tensorflow as tf
def diy_to_tensor(x):
    x = np.asarray(x).astype('float32')
    x = tf.convert_to_tensor(x)
    return x
#featurelist_NN = ['Power_kW', 'temp_C', 'windGust_m/s', 'Holiday', 'Weekday', 'Hour', 'Power -1']
featurelist_NN = ['Power_kW', 'pres_mbar', 'temp_C',
       'windGust_m/s', 'Holiday', 'Weekday', 'Hour', 'Month',
       'Power -1']
X_train, X_val, Y_train, Y_val, t_train, t_val = split_features(featurelist_NN)
X_train = diy_to_tensor(X_train)
X_val = diy_to_tensor(X_val)
Y_train = diy_to_tensor(Y_train)
Y_val = diy_to_tensor(Y_val)
X_train.shape
```

```python
import tensorflow as tf 

tfkl = tf.keras.layers
inp = tf.keras.Input(shape=(len(featurelist_NN)-1))
#x = tfkl.LSTM(16)(inp)
x = tfkl.Dense(16, activation="linear")(inp)
x = tfkl.Dense(16, activation="linear")(x)
x = tfkl.Dense(16, activation="linear")(x)
#x = tfkl.SimpleRNN(16, activation="linear")(x)
x = tfkl.Dense(1, activation="linear")(x)

model = tf.keras.models.Model(inp, x)
model.compile(loss="mae", optimizer="adam", metrics=["mae"])
model.summary()
```

```python
history = model.fit(X_train, Y_train,
          batch_size=256,
          epochs=10,
          verbose=1,
          validation_data=(X_val, Y_val))
```

```python
print(history.history.keys())
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss = mae')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
model.evaluate(X_val,  Y_val, verbose=2)
NN_val_pred = model.predict(X_val)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)

plt.plot(t_val,Y_val)
plt.plot(t_val,NN_val_pred)

plt.subplot(1,2,2)
plt.scatter(Y_val,NN_val_pred)

plt.show()

ml_metrics(NN_val_pred,Y_val)
```

Probably the neural network can perform better with more data and LSTM-Cell layers.
